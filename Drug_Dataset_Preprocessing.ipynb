#**Data Preprocessing**

Grouping Date and Removing Date Type

```python
ddeath.drop(['Date Type'], axis=1, inplace=True)
```

```python
ddeath['Date'] = pd.to_datetime(ddeath['Date'])
```

```python
# Change Date Values to Year and rename to YOD (year of death)
ddeath['YOD'] = ddeath['Date'].dt.year
ddeath.drop(['Date'], axis=1, inplace=True)
ddeath.head(5)
```

**Grouping and Encoding Age**

```python
minimum_age= drug['Age'].min()
maximum_age= drug['Age'].max()
print("Minimum Age:", minimum_age)
print("Maximum Age:", maximum_age)
```

```python
# Define bins
bins = list(range(10, 100, 10))  # Creates bins from 10 to 90 with 10-year intervals
labels = [f"{bins[i]} - {bins[i+1]-1}" for i in range(len(bins)-1)]  # Labels for groups

# Create a new column with age groups
ddeath['Age Group'] = pd.cut(ddeath['Age'], bins=bins, labels=labels, right=True)
# Display grouped counts
print(ddeath['Age Group'].value_counts().sort_index())
```

```python
# Age Group Encoding
# Create a mapping for encoding
age_encoding = {label: idx + 1 for idx, label in enumerate(labels)}
# Encode Age Group column
ddeath['Age Group Encoded'] = ddeath['Age Group'].map(age_encoding)
# Drop other age columns
ddeath = ddeath.drop(columns=['Age Group'], errors='ignore') # keep Age column for Transformation
# Display encoded values count
print(ddeath[['Age Group Encoded']].value_counts().sort_index())
```

**Encoding Sex (Binary Encoding)**

```python
# 'Male' will be 1, 'Female' will be 0, Null values will be -1
if 'Sex' in ddeath.columns:
    ddeath['Sex'] = (
        ddeath['Sex']
        .astype(str).str.strip()
        .map({'Male': 1, 'Female': 0})
        .fillna(-1)  # Fill NaNs with a placeholder
        .astype(int)
    )
ddeath.head(5)
```

**Latitude and Longitude**

```python
# Function to extract latitude and longitude
def extract_lat_lon(geo_value):
    if pd.isna(geo_value):
        return None, None  # Keep NaN values as they are
    try:
        lat_lon = geo_value.split('\n')[-1]  # Extract last part containing coordinates
        lat, lon = lat_lon.strip("()").split(",")  # Remove parentheses and split
        return float(lat.strip()), float(lon.strip())  # Convert to float
    except:
        return None, None  # Return None if format is incorrect
geo_columns = ['ResidenceCityGeo', 'InjuryCityGeo', 'DeathCityGeo']

# Apply function and create new LAT & LON columns
for col in geo_columns:
    lat_col = col.replace("Geo", "LAT")
    lon_col = col.replace("Geo", "LON")
    ddeath[lat_col], ddeath[lon_col] = zip(*ddeath[col].apply(extract_lat_lon))



# Display updated DataFrame with new LAT and LON columns
print(ddeath[['ResidenceCityLAT', 'ResidenceCityLON', 'InjuryCityLAT', 'InjuryCityLON', 'DeathCityLAT', 'DeathCityLON']].head())
```

```python
#Filling Missing Values
# Residence City, Death City, ResidenceCityGeo, InjuryCityGeo, ResidenceCityLAT, ResidenceCityLON, InjuryCityLAT, InjuryCityLON

# Fill missing values in 'Residence City', 'Death City' with the mode
for col in ['Residence City', 'Death City']:
    ddeath[col] = ddeath[col].fillna(ddeath[col].mode()[0])

# Fill missing geospatial data using the mean of the existing coordinates for the respective cities
for col in ['ResidenceCityGeo', 'InjuryCityGeo']:
    ddeath[col] = ddeath[col].fillna(ddeath[col].mode()[0])

# Fill missing latitude and longitude values with the mean of the existing coordinates
for col in ['ResidenceCityLAT', 'ResidenceCityLON', 'InjuryCityLAT', 'InjuryCityLON', 'DeathCityLAT', 'DeathCityLON']:
    ddeath[col] = ddeath[col].fillna(ddeath[col].mean())
```
